```
docA = 'the cat sat on my face'
docB = 'the dog sat on my bed'
```

```
bowA = docA.split(" ")
bowB = docB.split(" ")
```

```
wordSet = set(bowA).union(set(bowB))
```

```
wordSet
```

```
wordDictA = dict.fromkeys(wordSet,0)
wordDictB = dict.fromkeys(wordSet,0)
```

```
for word in bowA:
  wordDictA[word]+=1
for word in bowB:
  wordDictB[word]+=1
```

```
import pandas as pd
```

```
pd.DataFrame([wordDictA,wordDictB])
```

```
def computeTF(wordDict,bow):
  #term frequency
  tfDict = {}
  #total # words in bag
  bowCount = len(bow)
  print(bowCount)
  #iteritems lihaile
  for word,count in wordDict.iteritems():
    print(count)
    # set delimitor is float is very important, otherwise, it is integer. you can only get result of 0!!!
    tfDict[word] = count/float(bowCount)
  return tfDict
```

```
tfBowA = computeTF(wordDictA,bowA)
tfBowB = computeTF(wordDictB,bowB)
tfBowA
```

```
def computeIDF(docList):
  import math
  idfDict = {}
  N = len(docList)
  
  idfDict = dict.fromkeys(docList[0].keys(),0)
  for doc in docList:
    for word,val in doc.iteritems():
      if val > 0:
        idfDict[word] += 1
  
  for word,val in idfDict.iteritems():
    idfDict[word] = math.log(N/float(val))
  return idfDict
```

```
idfs = computeIDF([wordDictA,wordDictB])
print([wordDictA,wordDictB])
```

```
def computeTFIDF(tfBow,idfs):
  tfidf = {}
  for word, val in tfBow.iteritems():
    tfidf[word] = val*idfs[word]
  return tfidf
```

```
tfidfBowA = computeTFIDF(tfBowA,idfs)
tfidfBowB = computeTFIDF(tfBowB,idfs)
tfidfBowA
```
